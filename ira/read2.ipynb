```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

#  pkl 
pkl_path = "/home/yvonne/Documents/EyeBlink/ira_data/close_glass_20250329_144543.pkl"

#  pkl 
with open(pkl_path, 'rb') as f:
    temperature_frames = pickle.load(f)

# 
fps_values = [6.43, 6.47, 6.50, 6.53, 6.56]
avg_fps = sum(fps_values) / len(fps_values)  # 
time_interval_ms = 1000 / avg_fps            #   153.85 ms

# 
avg_temp_per_frame = []
for frame in temperature_frames[400:700]:
    center_region = frame[8:17, 10:22]  #  9x12 
    avg_temp = np.mean(center_region[center_region > 0])
    avg_temp_per_frame.append(avg_temp)

# 
start_frame = 0
num_frames = len(avg_temp_per_frame)
start_time_ms = start_frame * time_interval_ms
time_axis = np.arange(start_time_ms, start_time_ms + num_frames * time_interval_ms, time_interval_ms)

# 
plt.figure(figsize=(10, 4))
plt.plot(time_axis, avg_temp_per_frame, marker='o', linestyle='-', color='blue')
plt.title("Average Temperature of Center 9x12 Region over Time")
plt.xlabel("Time (ms)")
plt.ylabel("Average Temperature (C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
