```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# 替换为你真实的 pkl 文件路径
pkl_path = "ira_data/close_1cm_20250327_212657.pkl"

# 读取 pkl 文件
with open(pkl_path, 'rb') as f:
    temperature_frames = pickle.load(f)


# 提取中间区域并计算平均温度
avg_temp_per_frame = []

for frame in temperature_frames[300:600]:
    # 中心区域：行 4~18（共15行），列 6~26（共20列）
    center_region = frame[4:19, 6:26]
    avg_temp = np.mean(center_region[center_region > 0])
    avg_temp_per_frame.append(avg_temp)

# 画图
plt.figure(figsize=(10, 4))
plt.plot(avg_temp_per_frame, marker='o', linestyle='-', color='blue')
plt.title("Average Temperature of Center 15x20 Region per Frame")
plt.xlabel("Frame Number")
plt.ylabel("Average Temperature (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
